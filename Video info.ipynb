```python
import urllib.request
# open a connection to a URL using urllib
webUrl  = urllib.request.urlopen('https://www.youtube.com/watch?v=rfscVS0vtbw')

#get the result code and print it
print ("result code: " + str(webUrl.getcode()))    # result code : 200 , means the HTTP request is processed successfully

# read the data from the URL and print it
data = webUrl.read()
print (data)
```

```python
import numpy as np
import cv2 as cv
cap = cv.VideoCapture(r'C:\\Users\\user\\Pictures\\Video Projects\\20220408_202042.mp4')
if not cap.isOpened():
    print("Cannot open camera")
    exit()
while True:
    
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # if frame is read correctly ret is True
    if not ret :
        print("Can't receive frame (stream end?). Exiting ...")
        break
        
    # Our operations on the frame come here
    resized_frame = cv.resize(frame, (1000,600))
    
    cv.imshow('video', resized_frame)
    
    if cv.waitKey(7) == ord('q'):
        break
# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
```
